```python
import numpy as np
import pandas as pd
import cv2
import kagglehub
import os
```

```python
path = kagglehub.dataset_download("rahmasleam/flowers-dataset")
```

```python
 # prints the contents in that folder
folders = os.listdir(r"C:\Users\user\.cache\kagglehub\datasets\rahmasleam\flowers-dataset\versions\1\flower_photos")
folders
```

```python
features = []
class_labels = []
for folder in folders:
    for images in os.listdir(r"C:\Users\user\.cache\kagglehub\datasets\rahmasleam\flowers-dataset\versions\1\flower_photos\{}".format(folder)):
        img = cv2.imread(r"C:\Users\user\.cache\kagglehub\datasets\rahmasleam\flowers-dataset\versions\1\flower_photos\{}\{}".format(folder, images), 0) # Array representation # 0 --> Grayscale, 1 --> RGB image
        img = cv2.resize(img, (50,50)) # resizing the image
        img = img.flatten() # flattening the image
        features.append(img) # Appending each flattened image to list
        class_labels.append(folder) # Appending each class label to list
```

```python
final_data = pd.DataFrame(features)
final_data
```

```python
final_data["Class Labels"] = class_labels
final_data
```
