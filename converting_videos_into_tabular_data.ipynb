# Converting videos into multiple frames for tabular data
Use pixabay website for video dataset

#### Creating 3 different folders

```python
os.makedirs(r"C:\Users\user\ML\Week 2\dog_cat_&_elephant videos\dog")
os.makedirs(r"C:\Users\user\ML\Week 2\dog_cat_&_elephant videos\cat")
os.makedirs(r"C:\Users\user\ML\Week 2\dog_cat_&_elephant videos\elephant")
```

**By using the below code, we are extracting only the mp4 files**

```python
l = []
for y in os.listdir(r"C:\Users\user\ML\Week 2\dog_cat_&_elephant videos"):
    if y.endswith(".mp4"):
        l.append(y)
l
```

##### We are splitting each video into different frame images and saving it into respective folders created earlier

```python
for video in l:
    vid = cv2.VideoCapture(r"C:\Users\user\ML\Week 2\dog_cat_&_elephant videos\{}".format(video)) 
    c = 0
    while True:
        res, img = vid.read() 
        if res == False: 
            break
        # we extract the video name from mp4 file using split
        cv2.imwrite(r"C:\Users\user\ML\Week 2\dog_cat_&_elephant videos\{}\{}{}.jpg".format(video.split(".")[0], video.split(".")[0], c), img)
        c += 1
        cv2.imshow("video", img)
        if cv2.waitKey(10) & 255 == ord("q"): 
            break
    cv2.destroyAllWindows()
```

##### Now we convert all the images in each folder into tabular data

```python
lst = []
class_labels = []
folders = os.listdir(r"C:\Users\user\ML\Week 2\dog_cat_&_elephant videos")
for y in folders:
    if y.endswith(".mp4"):
        lst.append(y)
for x in folders:
    if x not in lst:
        class_labels.append(x)
class_labels
```

```python
cat_features = []
for image in os.listdir(r"C:\Users\user\ML\Week 2\dog_cat_&_elephant videos\cat"):
    img = cv2.imread(r"C:\Users\user\ML\Week 2\dog_cat_&_elephant videos\cat\{}".format(image), 1) 
    img = cv2.resize(img, (50,50))
    img = img.flatten()
    cat_features.append(img)
cat_data = pd.DataFrame(cat_features)
cat_data["Class Labels"] = class_labels[0]
cat_data.to_csv("cat_data.csv")
cat_data
```

```python
dog_features = []
for image in os.listdir(r"C:\Users\user\ML\Week 2\dog_cat_&_elephant videos\dog"):
    img = cv2.imread(r"C:\Users\user\ML\Week 2\dog_cat_&_elephant videos\dog\{}".format(image), 1) 
    img = cv2.resize(img, (50,50))
    img = img.flatten()
    dog_features.append(img)
dog_data = pd.DataFrame(dog_features)
dog_data["Class Labels"] = class_labels[1]
dog_data.to_csv("dog_data.csv")
dog_data
```
